```python
import pandas as pd
df = pd.read_csv("peliculas_enriquecidas.csv")
df.head()
```

```python
# Construímos el fichero text_for_vector

# Convertimos campos de listas en texto
df['keywords_text'] = df['keywords_tfidf'].apply(lambda x: ' '.join(x) if isinstance(x, list) else '')
df['genres_text'] = df['genres'].apply(lambda x: ' '.join(x) if isinstance(x, list) else '')

# Convertimos campos numéricos y manejamos nulos
df['topic_id_str'] = df['topic_id'].astype(str)
df['main_country'] = df['main_country'].fillna('unknown')
df['director'] = df['director'].fillna('unknown')
df['title'] = df['title'].fillna('')

# Creamos columna final para vectorización
df['text_for_vector'] = (
    df['overview_lemmatized'] + ' ' +
    df['keywords_text'] + ' ' +
    df['genres_text'] + ' ' +
    df['director'] + ' ' +
    df['main_country'] + ' ' +
    df['topic_id_str'] + ' ' +
    df['title']
)
```

```python
df.head()
```

```python
# Importamos vectorizador

from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(
    max_features=5000,
    stop_words='english',
    ngram_range=(1,2)
)
```

```python
tfidf_matrix = vectorizer.fit_transform(df['text_for_vector'])
```

```python
# Guardamos el vectorizador para usarlo después
import joblib
joblib.dump(vectorizer, 'tfidf_vectorizer.pkl')
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Matriz de similitudes: Cada fila contiene la similitud de una película con todas las demás
similarity_matrix = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
# Creamos función de recomendación

df = pd.read_csv('peliculas_enriquecidas.csv')

# Reseteamos índice
df = df.reset_index(drop=True)

# Mapa titulo -> indice
title_to_index = pd.Series(df.index, index=df['title'])

def recomendar_peliculas(titulo, top_n=5):
  if titulo not in title_to_index:
    print(f"Titulo '{titulo}' no encontrado.")
    return []
  # Indice de la película base
  idx = title_to_index[titulo]

  # Similarización de esta película con todas
  sim_scores = list(enumerate(similarity_matrix[idx]))

  # Ordenar por similitud excepto consigo misma
  sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
  sim_scores = sim_scores[1:top_n+1]

  # Obtener títulos de películas similares
  peliculas_similares = [(df.iloc[i]['title'], df.iloc[i]['id'], round(score, 3)) for i, score in sim_scores]

  return peliculas_similares
```

```python
# Probamos

recomendaciones = recomendar_peliculas("Los aristogatos", top_n=5)
for titulo, id_peli, score in recomendaciones:
  print(f"{titulo} (ID: {id_peli}) - Similitud: {score}")
```
